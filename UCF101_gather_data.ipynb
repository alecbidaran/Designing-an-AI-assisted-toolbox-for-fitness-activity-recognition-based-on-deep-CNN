```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
def backsub(video_dir):
  print(f"__backubing{video_dir}___")
  capture = cv2.VideoCapture(video_dir)
  frame_count=int(capture.get(cv2.cv2.CAP_PROP_FRAME_COUNT))
  backsub=cv2.createBackgroundSubtractorMOG2()
  frames=[]
  for i in range(10):
    ret,frame=capture.read()
    frame=cv2.resize(frame,(128,128),cv2.INTER_AREA)
    back=backsub.apply(frame)
    ret,back=cv2.threshold(back,150,255,cv2.THRESH_BINARY)
    frames.append(back)
  return np.array(frames)
    
```

```python
def get_frame(video_dir):
  capture = cv2.VideoCapture(video_dir)
  frame_count=int(capture.get(cv2.cv2.CAP_PROP_FRAME_COUNT))
  frames=[]
  for i in range(10):
    ret,frame=capture.read()
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    frame=cv2.resize(frame,(128,128),cv2.INTER_AREA)
    #back=backsub.apply(frame)
    #ret,back=cv2.threshold(back,150,255,cv2.THRESH_BINARY)
    frames.append(frame)
  return np.array(frames,dtype=np.int8)
```

# Create Pytorch Dataset

```python
bench_list=glob.glob("/content/Fitness_UCF/Benchpress/*")
push_list=glob.glob("/content/Fitness_UCF/Pushups/*")
pullups=glob.glob("/content/Fitness_UCF/Pullups/*")
squat=glob.glob("/content/Fitness_UCF/Squat/*")
```

```python
## meta 
def generate_dataframe(label,dir):
  labels=[label for _ in range(len(dir))]
  return pd.DataFrame([dir,labels]).T

bench_d=generate_dataframe("benchpress",bench_list)
push_d=generate_dataframe("pushups",push_list)
pull_d=generate_dataframe("pullups",pullups)
squat_d=generate_dataframe("squats",squat)
metadata=pd.concat([bench_d,push_d,pull_d,squat_d])
metadata.columns=["dir","label"]
metadata.to_csv("metadata.csv",index=False)
```

```python
from torch_snippets import * 
import torch 
```

```python
mt=pd.read_csv("metadata.csv")
mt
```

```python
label_to_target={l:t for t,l in enumerate(mt["label"].unique())}
label_to_target
```

```python
class fitness_dataset(torch.utils.data.Dataset):
  def __init__(self,metadata):
    self.metadata=metadata
  def __getitem__(self,idx):
    video_dir=self.metadata["dir"][idx]
    label_name=self.metadata["label"][idx]
    frames=get_frame(video_dir)
    frames=torch.Tensor(frames).float().permute(0,3,1,2)
    label=torch.Tensor(label_to_target[label_name]).long()
    return frames,label
  def collate_fn(self,batch):
    return tuple(zip(*batch))
    
```

```python
fit_ds=fitness_dataset(mt)
```
